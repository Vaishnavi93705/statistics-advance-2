1.What is hypothesis testing in statistics?
- Hypothesis testing in statistics is a method used to make decisions or draw conclusions about a population based on sample data.

  It provides a structured way to test assumptions (claims) about a parameter
  (like a mean, proportion, or variance).


2.What is the null hypothesis, and how does it differ from the alternative
hypothesis?
- The null hypothesis is a formal statement used in hypothesis testing which assumes that there is no significant difference or effect in the population parameter being studied

  The alternative hypothesis is the contradictory statement to the null hypothesis. It suggests that there is a significant difference or effect in the population parameter. It reflects what the researcher aims to prove.

3.Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.

- The significance level, denoted by α, is a threshold set by the researcher before conducting the test.It represents the maximum probability of making a Type I error (rejecting the null hypothesis when it is actually true).

  Role in Hypothesis Testing

     - Comparison with p-value
     - Defining the Critical Region

4.What are Type I and Type II errors? Give examples of each.
- Type I Error (False Positive)

  - Occurs when we reject the null hypothesis (H₀) even though it is actually true.In other words, we conclude there is an effect or difference when in reality, there isn’t.Probability of making this error = α (significance level).
  - Example:
  If we set α = 0.05, then there is a 5% chance of rejecting a true null hypothesis.

    H₀: μ = 50 (the population mean is 50)

    If H₀ is actually true but we conclude μ ≠ 50 → Type I error.

- Type II Error (False Negative)

   - Occurs when we fail to reject the null hypothesis (H₀) even though it is actually false.In other words, we conclude there is no effect or difference when in reality, there is.Probability of making this error = β (related to the test’s power = 1 − β).

   - H₀: μ = 50 (the population mean is 50)

     Suppose the true mean is actually μ = 55.

      If the test fails to detect this difference and we accept H₀ → Type II error.

5.What is the difference between a Z-test and a T-test? Explain when to use
each.

- Z-test

  - A parametric test used when the population variance (σ²) is known or the sample size is large (n > 30).

  - Based on the standard normal distribution (Z-distribution), which has mean = 0 and standard deviation = 1.

  - When to use a Z-test:

    Population variance (σ²) is known.

    Large sample size (n > 30).

    Data approximately normally distributed.

- T-test

   - A parametric test used when the population variance (σ²) is unknown and must be estimated from the sample.

   - Based on the t-distribution, which is similar to the normal distribution but has heavier tails (to account for extra uncertainty with small samples).

   - When to use a T-test:

      Population variance is unknown.

      Small sample size (n ≤ 30).

      Data approximately normally distributed.



6.Write a Python program to generate a binomial distribution with n=10 and
p=0.5, then plot its histogram

```python
import numpy as np
import matplotlib.pyplot as plt

n = 10
p = 0.5
size = 1000

data = np.random.binomial(n, p, size)


plt.hist(data, bins=np.arange(0, n+2) - 0.5, edgecolor='black', density=True)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.show()
```

```python
import numpy as np
from scipy.stats import norm

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]


mu_0 = 50
alpha = 0.05


sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=0)
n = len(sample_data)


z_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print("Sample Mean:", sample_mean)
print("Sample Std Dev:", sample_std)
print("Z-statistic:", z_stat)
print("p-value:", p_value)

if p_value < alpha:
    print("Reject the null hypothesis (significant difference).")
else:
    print("Fail to reject the null hypothesis (no significant difference).")
```

8.: Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


np.random.seed(42)
mu, sigma = 50, 5
n = 100

data = np.random.normal(mu, sigma, n)

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)


alpha = 0.05
z_critical = norm.ppf(1 - alpha/2)
margin_of_error = z_critical * (sample_std / np.sqrt(n))

ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print("Sample Mean:", sample_mean)
print("95% Confidence Interval: (", ci_lower, ",", ci_upper, ")")


plt.hist(data, bins=15, edgecolor='black', alpha=0.7, density=True)
plt.title("Simulated Normal Distribution Data")
plt.xlabel("Value")
plt.ylabel("Density")


xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, sample_mean, sample_std)
plt.plot(x, p, 'k', linewidth=2)

plt.show()
```

9.Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_zscores(data):

    data = np.array(data)


    mean = np.mean(data)
    std_dev = np.std(data)

    z_scores = (data - mean) / std_dev


    print("Original Mean:", mean)
    print("Original Std Dev:", std_dev)
    print("First 5 Z-scores:", z_scores[:5])


    plt.hist(z_scores, bins=15, edgecolor='black', alpha=0.7, density=True)
    plt.title("Histogram of Standardized Data (Z-scores)")
    plt.xlabel("Z-score")
    plt.ylabel("Density")
    plt.axvline(0, color='red', linestyle='--', label="Mean (Z=0)")
    plt.legend()
    plt.show()

    return z_scores

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

calculate_zscores(sample_data)
```
